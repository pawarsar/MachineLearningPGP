```python
def lengthOfLongestSubstring(s):
    """
    :type s: str
    :rtype: int
    """
    ss = s.split()
    prev_ss = ss[0]#'a'
    ss_value = prev_ss
    for substring in ss[1:]:#['a','b','c','b','c','b','b']
        if substring in ss_value:
            ss_value = prev_ss
            print("if", ss_value)
        else:
            ss_value += substring
            print("else",ss_value)
        prev_ss = substring
        print("prevss",prev_ss)
    
    return len(ss_value)
```
