```python
# to load and manipulate data
import pandas as pd
import numpy as np

# to visualize data
import matplotlib.pyplot as plt
import seaborn as sns

# to split the data into train and test sets
from sklearn.model_selection import train_test_split

# to build a linear regression model
from sklearn.linear_model import LinearRegression

# to check a regression model's performance
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
df = pd.read_csv("backpain.csv")
data = df
```

```python
# Load the dataset and identify the variables that have a correlation greater than or equal to 0.7 with the ‘pelvic_incidence’ variable.
correlation_matrix = df.corr(numeric_only=True)
pelvic_incidence_corr = correlation_matrix['pelvic_incidence']
highly_correlated_variables = pelvic_incidence_corr[abs(pelvic_incidence_corr) >= 0.7].index.tolist()
print("Variables with correlation >= 0.7 with 'pelvic_incidence':")
print(highly_correlated_variables)
```

```python
# Encode Status variable: Abnormal class to 1 and Normal to 0.

# Split the data into a 70:30 ratio. What is the percentage of 0 and 1 classes in the test data (y_test)?

# Note - Do not use stratify on the dataset.

# Encode the 'Status' variable
df['Status'] = df['Status'].map({'Abnormal': 1, 'Normal': 0})
# Split the data into features and target variable
X = df.drop('Status', axis=1)
y = df['Status']
# Split the data into training and testing sets (70:30 ratio)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
# Calculate the percentage of 0 and 1 classes in the test data
class_counts = y_test.value_counts(normalize=True) * 100
print("Percentage of classes in the test data:")
print(class_counts)
```

```python
# Which metric is the most appropriate metric to evaluate the model according to the problem statement?
# Accuracy
# Recall
# Precision
# F1 score
# Since this is a binary classification problem, the most appropriate metrics to evaluate the model would be:
# - Accuracy: to measure the overall correctness of the model.
# - Precision: to measure the proportion of true positive predictions among all positive predictions.
# - Recall: to measure the proportion of true positive predictions among all actual positives.
# - F1 Score: to provide a balance between precision and recall, especially when the classes are imbalanced.
# Build a logistic regression model to predict the ‘Status’ variable using the variables identified in step 1 as features. Evaluate the model using the appropriate metrics identified in step 3.
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
# Select the features identified in step 1
features = highly_correlated_variables
X_train_selected = X_train[features]
X_test_selected = X_test[features]
# Build the logistic regression model
model = LogisticRegression()
model.fit(X_train_selected, y_train)
# Make predictions on the test set
y_pred = model.predict(X_test_selected)
# Evaluate the model using the appropriate metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
print("Model Evaluation Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")

```
